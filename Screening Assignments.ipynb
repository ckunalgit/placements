## Screening Assignment 01

```python
class screen01:
    
    def Solution(self, filename, text1, text2):
        
        '''This function reads the given text file and replaces specific text with another text'''
        
        # Read each line from the file and create a list of all words
        with open(filename, encoding = 'utf-8') as f:
            lines = f.readlines()
            for i in lines:
                mt = i.split(' ')
        
        # Go through the list created above and replace the text and write back into the list
        for i,j in enumerate(mt):
            if j == text1:
                mt[i] = text2
            else:
                continue
        
        # Write back the contents of the list into the file
        with open("example.txt", 'w', encoding = 'utf-8') as f:
            for i in mt:
                f.write(i + ' ')
```

```python
obj01 = screen01()
```

```python
obj01.Solution('example.txt', 'placement', 'screening')
```

##### A decorator class in python is used to modify the behaviour of a function or a class

```python
# defining a decorator
def demo_decor(fun):

    def inner1():
        print("Hello, this is before function execution")

        fun()

        print("This is after function execution")
    return inner1


# defining a function, to be called inside wrapper
def function_to_be_used():
    print("This is inside the function !!")


# passing 'function_to_be_used' inside the
# decorator to control its behaviour
function_to_be_used = demo_decor(function_to_be_used)


# calling the function
function_to_be_used()
```

### Inheritance can be of the following types:
    1. Single level : Only one child which inherits one parents
    2. Multi level: Grandparent, parent and child class which goes on...
    3. Multiple : Multiple child classes inherit one single parent class

### Multi Level Inheritance

```python
class grandParent:
    
    def func1(self):
        print("This is grandparent method")

class parent(grandParent):
    
    def func2(self):
        print("This is grandparent method")
        
class child(parent):
    
    def func3(self):
        print("This is grandparent method")
                
```

```python
# Creating an object of child class
obj = child()
```

##### See that all the methods of the parent and grandparent class are available to the child

#### Multiple inheritance and Method Resolution Order (MRO)

```python
class dad:
    
    def __init__(self):
        print("This is dad's init")
        
    def func1(self):
        print("This is dad's method")

class mom:
    
    def __init__(self):
        print("This is mom's init")
    
    def func1(self):
        print("This is mom's method")
        
# Child class inherits mom and dad
class child(dad,mom):
    
    def __init__(self):
        print("This is child init")
    
    def func1(self):
        print("This is child method")
                
```

#### As an object is initialized the child class init will be printed

```python
obj = child()
```

```python
class dad:
    
    def __init__(self):
        print("This is dad's init")
        
    def func1(self):
        print("This is dad's method")

class mom:
    
    def __init__(self):
        print("This is mom's init")
    
    def func1(self):
        print("This is mom's method")
        
# Child class inherits mom and dad
class child(dad,mom):
    
    def __init__(self):
        print("This is child init")
        
        super().__init__()                 ## This displays MRO
    
    def func1(self):
        print("This is child method")
        
        super().func1()                    ## MRO use on methods
                
```

##### In the above operation, we have called the 'Super' constructor, but since there are 2 parent classes, Python uses left -> right order

```python
obj = child()
```

```python
# This also applies to methods of the parent class

obj.func1()
```
